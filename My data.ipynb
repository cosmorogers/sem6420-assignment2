```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/train_parkinsons.csv')

df = df.drop(["MDVP_Fhi","MDVP_Jitter_pct","MDVP_Jitter_abs","MDVP_PPQ","Jitter_DDP","MDVP_Shimmer",
              "MDVP_Shimmer_db","Shimmer_APQ3","Shimmer_APQ5","Shimmer_DDA","NHR","HNR","RPDE","DFA","spread2",
              "D2","PPE"], axis=1)

df.info()
```

```python
cols = df.columns.tolist()
print cols
cols = [cols[5]] + cols[0:5]
df = df[cols]

print(df.columns.tolist())
```

```python
df.head(10)
```

```python
# %matplotlib inline: To make matplotlib inline graphics
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
scale        = lambda x: (x - x.mean()) / x.std()
from sklearn.decomposition import PCA

# use .iloc to get subset of DataFrame
df_std = df.iloc[0:,2:].apply(scale)

pca   = PCA(n_components = 4).fit(df_std)
vr = pca.explained_variance_ratio_
print("Variance exlained : %0.3f, %0.3f, %0.3f, %0.3f" % (vr[0], vr[1], vr[2], vr[3]))
print("Total variance explained with the first 4 PCs: %0.3f" % (sum(vr)))
print(pca.components_)


Xtrans = pca.transform(df_std)
dfx = pd.DataFrame(Xtrans, columns=['PC1', 'PC2', 'PC3', 'PC4'])
```

```python
plt.figure()
ax=dfx[df['status']==1].plot(kind='scatter', x='PC1', y='PC2', color='red', label='Parkinson')
dfx[df['status']==0].plot(kind='scatter', x='PC1', y='PC2', color='green', label='Healthy ',ax=ax)
plt.title('PCA of Parkinson training dataset')

ax=dfx[df['status']==1].plot(kind='scatter', x='PC1', y='PC3', color='red', label='Parkinson')
dfx[df['status']==0].plot(kind='scatter', x='PC1', y='PC3', color='green', label='Healthy ',ax=ax)
plt.title('PCA of Parkinson training dataset')

ax=dfx[df['status']==1].plot(kind='scatter', x='PC2', y='PC3', color='red', label='Parkinson')
dfx[df['status']==0].plot(kind='scatter', x='PC2', y='PC3', color='green', label='Healthy ',ax=ax)
plt.title('PCA of Parkinson training dataset')
```

```python
#Adapted from http://scikit-learn.org/stable/auto_examples/decomposition/plot_pca_vs_lda.html
X = df[df.columns[1:6]].values
y = df[df.columns[0]].values

target_names = ['Healthy', 'Parkinson']

from sklearn.lda import LDA

lda = LDA(n_components=2)
X_r2 = lda.fit(X, y).transform(X)
#X_r2 = lda.fit(df_std.values, y).transform(df_std) #get the same thing


plt.figure()
for c, i, target_name in zip("gr", [0,1], target_names):
    #plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], color=c, label=target_name) #index error
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 0], color=c, label=target_name)
plt.legend()
plt.title('LDA of dataset')
```

```python
from sklearn.cross_validation import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.metrics import roc_curve, auc

train_data = df.values
X = train_data[:, 2:]
y = train_data[:, 0]

n = len(df)/2

cv = KFold(n=len(train_data), n_folds=5)
```

```python
mean_acc = 0.0
mean_auc = 0.0
all_tpr = []
all_acc = []
all_auc = []
i=0

plt.figure(figsize=(8,8))


for training_set, test_set in cv:
    X_train = X[training_set]
    y_train = y[training_set]
    X_test = X[test_set]
    y_test = y[test_set]
    model = RandomForestClassifier(n_estimators=100)
    model.fit(X_train, y_train)
    y_prediction = model.predict(X_test)    
    
    # Get probability output from the model
    y_prob = model.predict_proba(X_test)
    acc = np.sum(y_test == y_prediction)*1./len(y_test)
    print "Prediction accuracy:", acc
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(y_test, y_prob[:, 1])
    roc_auc = auc(fpr, tpr)
    print "Area under ROC curve (AUC):", roc_auc
    plt.plot(fpr, tpr, lw=1, label='ROC fold %d (area = %0.2f)' % (i, roc_auc))    
    all_acc.append(acc)
    all_auc.append(roc_auc)
    
plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6))#, label='Luck')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Random Forest (Dataset B)')
plt.legend(loc="lower right")
plt.show()

all_acc=np.asarray(all_acc)
all_auc=np.asarray(all_auc)
print(all_acc)
# print 95% C.I. for both accuracy and AUC based on CV
print("Mean Accuracy: %0.3f (+/- %0.3f)" % (all_acc.mean(), all_acc.std() * 1.96))
print(all_auc)
print("Mean AUC: %0.3f (+/- %0.3f)" % (all_auc.mean(), all_auc.std() * 1.96))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

rng = np.random.RandomState(1)

mean_acc = 0.0
mean_auc = 0.0
all_tpr = []
all_acc = []
all_auc = []
i=0

plt.figure(figsize=(8,8))


for training_set, test_set in cv:
    X_train = X[training_set]
    y_train = y[training_set]
    X_test = X[test_set]
    y_test = y[test_set]
    model = AdaBoostClassifier(DecisionTreeClassifier(max_depth=4),
                          n_estimators=300, random_state=rng)
    model.fit(X_train, y_train)
    y_prediction = model.predict(X_test)    
    
    # Get probability output from the model
    y_prob = model.predict_proba(X_test)
    acc = np.sum(y_test == y_prediction)*1./len(y_test)
    print "Prediction accuracy:", acc
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(y_test, y_prob[:, 1])
    roc_auc = auc(fpr, tpr)
    print "Area under ROC curve (AUC):", roc_auc
    plt.plot(fpr, tpr, lw=1, label='ROC fold %d (area = %0.2f)' % (i, roc_auc))    
    all_acc.append(acc)
    all_auc.append(roc_auc)
    i = i+1
    
plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6))#, label='Luck')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC AdaBoost Decision Tree (Dataset B)')
plt.legend(loc="lower right")
plt.show()

all_acc=np.asarray(all_acc)
all_auc=np.asarray(all_auc)
print(all_acc)
# print 95% C.I. for both accuracy and AUC based on CV
print("Mean Accuracy: %0.3f (+/- %0.3f)" % (all_acc.mean(), all_acc.std() * 1.96))
print(all_auc)
print("Mean AUC: %0.3f (+/- %0.3f)" % (all_auc.mean(), all_auc.std() * 1.96))
```
